```python
import pandas as pd
data = pd.read_csv(r"/content/laptop.csv")
data.head(3)
```

```python
data.shape
```

```python
#nulls in the data
data[data.isnull().any(axis =1)]
```

```python
#To remove the nulls and '?' in the inches column
data["Inches"] = pd.to_numeric(data["Inches"],errors = 'coerce')
data = data.dropna(subset=["Inches"])
```

```python
#Checking for the nulls
data.isnull().sum()
```

```python
#Replacing the text in ram column to make it numeric
data["Ram"] = data["Ram"].str.replace("GB","").astype(int)
data.head(2)
```

```python
#To split the memory column to memory and drive columns
data[["Memory","Drive"]] = data["Memory"].str.split(" ",n=1,expand=True)
data.head(2)
```

```python
#Converting TB to GB and removing them from memory column by defining a function
def conversion(val):
  val = str(val).upper().strip()
  if "TB" in val:
    num = float(val.replace("TB",""))
    return num*1000
  elif "GB" in val:
    num=float(val.replace("GB",""))
    return num
  else:
    return None
```

```python
#Applying the function to the memory column
data["Memory_gb"] = data["Memory"].apply(conversion)
```

```python
#Checking nulls after splitting the columns
data[data.isnull().any(axis=1)]
```

```python
#Filling the null values median for numeric and mode for text data
data["Memory_gb"] = data["Memory_gb"].fillna(data["Memory_gb"].median())
data["Drive"] = data["Drive"].fillna(data["Drive"].mode()[0])
```

```python
#Dropping the columns not required for model
data = data.drop(["Unnamed: 0.1","Unnamed: 0","Company","TypeName","OpSys","Memory"],axis =1)
data.head(2)
```

```python
#Getting GHz value from the cpu column and converting it to numeric
data["GHZ"] = data['Cpu'].str.split(" ").str[-1]
data["GHZ"] = data["GHZ"].str.replace("GHz","").astype(float)
```

```python
#Defining a function to categorize the cpu column
def categorize(var):
  var= var.upper()
  if 'ATOM' in var or 'E-SERIES' in var:
        return 0
  elif 'I3' in var or 'A8' in var or 'A10' in var or 'RYZEN 3' in var:
        return 2
  elif 'I5' in var or 'RYZEN 5' in var:
        return 3
  elif 'I7' in var or 'I9' in var or 'RYZEN 7' in var or 'RYZEN 9' in var or 'XEON' in var:
        return 4
  else:
        return 2
```

```python
#applying the function on cpu column
data["Cpu"] = data["Cpu"].apply(categorize)
```

```python
#Defining a function to categorize the gpu column
def gpu_category(gpu):
  gpu = gpu.upper()
  if "HD" in gpu or "UHD" in gpu or "IRIS" in gpu or "R2" in gpu or "R3" in gpu or "R4" in gpu or "ARM" in gpu:
    return 1
  elif "MX" in gpu or "920" in gpu or "930" in gpu or "940" in gpu or "R5" in gpu or "R7 M3" in gpu:
    return 2
  elif "950" in gpu or "960" in gpu or "965" in gpu or "1050" in gpu or "RX 540" in gpu or "RX 550" in gpu or "RX 560" in gpu:
    return 3
  elif "1060" in gpu or "1070" in gpu or "1080" in gpu or "980" in gpu or "RX 570" in gpu or "RX 580" in gpu:
    return 4
  elif "QUADRO" in gpu or "FIREPRO" in gpu:
    return 5
  else:
    return 2
```

```python
#applying the function on cpu column
data["Gpu"] = data["Gpu"].apply(gpu_category)
```

```python
#Extracting the pixels from screenresolution
#Splitting pixels to length and width to categorize them
data["Pixels"] = data['ScreenResolution'].str.split(" ").str[-1]
data[["P_width","P_Height"]] = data["Pixels"].str.split("x",expand=True).astype(float)
```

```python
#Defining a pixel category function
def pixel_category(row):
  width = row["P_width"]
  height = row["P_Height"]
  pixels=width*height
  if pixels <= 1366*768:
    return 1
  elif pixels <= 1600*900:
    return 2
  elif pixels <= 1920*1200:
    return 3
  elif pixels <= 2880*1800:
    return 4
  else:
    return 5
```

```python
#Applying the function to resolution column
data["Resolution"] = data.apply(pixel_category,axis=1)
```

```python
#Creating a binary column for touchscreen laptops
data["Touchscreen"] = data["ScreenResolution"].str.contains("Touchscreen").astype(int)
```

```python
#Dropping the unnecessary columns
data = data.drop(["ScreenResolution","Weight","Pixels","P_width","P_Height"],axis=1)
data.head(3)
```

```python
#Defining a function for upper and lower boundaries of a box plot for outlier identification
def boundaries(col):
  q1 = col.quantile(0.25)
  q3 = col.quantile(0.75)
  iqr = q3-q1
  upper = q3 + 1.5*iqr
  lower = q1 - 1.5*iqr
  print("Upper_boundary -",upper)
  print("Lower_boundary -",lower)
```

# Training the models

```python
 #Standardizing the feature columns
 from sklearn.preprocessing import StandardScaler
 ss = StandardScaler()
 X = data.drop(["Price"],axis = 1)
 Y = data[["Price"]]
 X_scaled = ss.fit_transform(X)
```

```python
#Splitting the data into training and testing
from sklearn.model_selection import train_test_split
tts = train_test_split
x_train,x_test,y_train,y_test = tts(X_scaled,Y,test_size = 0.3,random_state = 45)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Training the linear model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Predicting the test values
test_predictions = lr.predict(x_test)
test_predictions[:5]
```

```python
#Predicting the train values
train_predictions = lr.predict(x_train)
train_predictions[:5]
```

```python
#A dataframe for the original and values predicted by linear model
result = pd.DataFrame(y_test)
result["linear_Predicted"] = test_predictions
result.head()
```

```python
#Errors in the predicted values
result["Linear_errors"] = result["Price"]-result["linear_Predicted"]
result.head()
```

```python
#R2_score of test data
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
test_score = r2_score(result["Price"],result["linear_Predicted"])
print("Test r2 score  ",round(test_score,3))
```

```python
#R2_score of trained data
train_score = r2_score(y_train,train_predictions)
print("Trained r2 score  ",round(train_score,3))
```

```python
#Mean absolute error, root mean squared error and mean squared errors of the linear model
lmae = mean_absolute_error(result["Price"],result["linear_Predicted"])
lmse = mean_squared_error(result["Price"],result["linear_Predicted"])
lrmse = root_mean_squared_error(result["Price"],result["linear_Predicted"])
print("Mean absolute error ",lmae)
print("Mean squared error ", lmse)
print("Root mean squared error ",lrmse)
```

```python
#Training the random forest regressor model
from math import sqrt
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=45,n_estimators = 400,max_depth = 5,min_samples_split=8, min_samples_leaf= 8,max_features="sqrt")
nrfr = rfr.fit(x_train,y_train)
nrfr
```

```python
#test data predictions of random forest regressor model
forest_test_predictions = rfr.predict(x_test)
forest_test_predictions[:5]
```

```python
#train data predictions of random forest regressor model
forest_train_predictions = rfr.predict(x_train)
forest_train_predictions[:5]
```

```python
#Adding the predicted values by random forest model to the result table
result["Forest_predicted"] = forest_test_predictions
result["forest_errors"] = result["Price"] - result["Forest_predicted"]
result.head()
```

```python
 #Adding the predicted values by random forest model to the result table
result["Forest_predicted"] = forest_test_predictions
result["forest_errors"] = result["Price"] - result["Forest_predicted"]
result.head()
```

```python
#r2 score of test data
forest_test_r2score = r2_score(result["Price"],result["Forest_predicted"])
round(forest_test_r2score,2)
```

```python
#r2 score of trained data
forest_train_r2score = r2_score(y_train,forest_train_predictions)
round(forest_train_r2score,2)
```

```python
#Mean absolute error, root mean squared error and mean squared errors of the random forest model
rmae = mean_absolute_error(result["Price"],result["Forest_predicted"])
rmse = mean_squared_error(result["Price"],result["Forest_predicted"])
rrmse = root_mean_squared_error(result["Price"],result["Forest_predicted"])
print("Mean absolute error ",rmae)
print("Mean squared error ", rmse)
print("Root mean squared error ",rrmse)
```
